```python
df["order_completed"] = df["order_completed"].str.replace(r'\+\d+', '', regex=True)
df["order_completed"] = pd.to_datetime(df["order_completed"], format="%Y-%m-%d %H:%M:%S.%f")
```

```python
df.dtypes
```

```python
df = df.astype({"user_id": float, "order_id": float})
```

```python
df.dtypes
```

```python
res_list = []
for i in range(len(df)):
    x = tuple(df.iloc[i])
    res_list.append(x)
res_list
```

```python
file = open("test_pokupon.txt", "w")

for tuple in res_list:
    file.write(str(tuple) + "," + "\n")
file.close()
```
